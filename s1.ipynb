Импортим импорты

```python
import zipfile as zf
import os
import pandas as pd
from collections import Counter
import time
from copy import deepcopy
```

Создаём директории

```python
#Папка для файлов из архива
if not os.path.exists("data/"):
    os.makedirs("data/")

#Папка для файлов, конвертированных в csv
if not os.path.exists("data/csv/"):
    os.makedirs("data/csv/")
```

Экстрактим данные, так быстрее чем ручками

```python
zip = "set.zip"
with zf.ZipFile(zip, 'r') as zip_ref:
    zip_ref.extractall("data/")
```

Получаем список с путями к каждому файлу

```python
files_only = os.listdir("data/set/")
files = []
for file in files_only:
    file = os.path.join("data/set/",file)
    files.append(file)
```

Функция для получения словаря со всеми возможными значениями колонок и списками в качестве значения

```python
def empty_dict_creator():
    keys_main = {}
    for idx, file in enumerate(files):
        df = pd.read_table(file, delimiter=",").drop(["Time"], axis=1)
        dict = df['Parameter'].value_counts().to_dict()
        keys_main = Counter(keys_main) + Counter(dict)
        single_df = df

    #Конвертируем результат в список для дальнейшей конвертации в словарь со списками
    keys_main = list(keys_main.keys())

    #Создаём словарь со списками для генерации строки
    empty_keys_list_dict = {}
    for i in keys_main:
        empty_keys_list_dict[i] = []
    
    return empty_keys_list_dict
```

```python
empty_dict_list = empty_dict_creator()
```

```python
empty_dict_list
```

Функция для генерации строки

```python
def row_from_dataframe(df):
    keys_list_dict = deepcopy(empty_dict_list)
    for idx, row_list in enumerate(range(df.shape[0])):
        row_list = df.loc[idx, :].values.flatten().tolist()
        keys_list_dict[row_list[0]].append(row_list[1])
    ret_dict = {}
    for i in keys_list_dict:
        try:
            ret_dict[i] = sum(keys_list_dict[i])/len(keys_list_dict[i])
        except:
            ret_dict[i] = -1
    return ret_dict    
```

Создадим итоговый словарь

```python
final_dict = deepcopy(empty_dict_list)
for idx, i in enumerate(files):
    start = time.time()
    df = pd.read_table(i, delimiter=",").drop(["Time"], axis=1)
    dict = row_from_dataframe(df)
    for i in final_dict:
        final_dict[i].append(dict[i])
    end = time.time()
    print(f"Затраченное время на добавление {idx} - {end - start}")
```

Создадим итоговый датафрейм

```python
final_df = pd.DataFrame(data=final_dict)
```

```python
final_df.to_csv("result.csv")
```
